Separo a música em intervalos de ,aproximadamente, um segundo.

```python
def divide(music,duration=60):
    y, fs = librosa.load(music,None, mono=True, duration = duration)
    y = y[20000:]
    z= []
    for i in range(0,len(y),20100):
        try:
            z.append(y[i:i+20100])
        except:
            z.append(y[i:])
            return [np.array(z),fs]
    return [np.array(z),fs]
```

Construo a cadeia de markov para cada intervalo.

```python
def markov(temp = 30):
    A = st.uniform.rvs(size = 4)
    A = A.reshape(2,2)
    for i in range(2):
        A[i][:] = A[i][:]/sum(A[i][:])
    nota = []
    A = np.matrix(A)
    print('Matriz de transição{}'.format(A))
    n_0 = np.array(random.choice(([[1],[0]],[[0],[1]])))
    n_0 = n_0.reshape(1,2)
    if n_0[0][0] < 1:
        nota.append(1)
    else:
        nota.append(0)
    for i in range(1,temp):
        n_0 = n_0*(A**i)
        if np.array(n_0)[0][0] < random.random():
            nota.append(0)
        else:
            nota.append(1)
    return nota
    
```

Dividindo as músicas

```python
h_d,fs_1 = divide('Guns.wav',duration = 30)
w_d,fs = divide('Moonlight.wav', duration = 30)
```

```python
k=[]
for i,p in enumerate(nota):
    if p == 0:
        k = k + list(h_d[i])
    else:
        k = k + list(w_d[i])
k = np.array(k)
```

```python
ipd.Audio(k, rate=fs)
```
